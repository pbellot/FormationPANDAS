## Réalisation de l'ACP

```python
from sklearn.decomposition import PCA
```

```python
acp = PCA()
```

```python
coordonnees = acp.fit_transform(st)
```

```python
acp.n_components_
```

```python
print(acp.explained_variance_)
```

```python
print(acp.explained_variance_ratio_)
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.plot(np.arange(1,acp.n_components_+1),acp.explained_variance_ratio_)
plt.title("Plot du % de variance expliquée")
plt.ylabel("% de variance")
plt.xlabel("Numéro du facteur")
plt.show()
```

```python
plt.plot(np.arange(1,acp.n_components_+1),np.cumsum(acp.explained_variance_ratio_))
plt.title("Plot du cumul de variance expliquée")
plt.ylabel("% de variance")
plt.xlabel("Numéro du facteur")
plt.ylim(0,1)
plt.show()
```

## Visualisation des variables

```python
print(acp.components_)
#chaque ligne = 1 facteur, chaque colonne = 1 variable d'origine
```

```python
valeurspropres = (df.shape[0]-1)/df.shape[0]*acp.explained_variance_
```

```python
sqrt_valeurspropres = np.sqrt(valeurspropres)
```

```python
st.shape
```

```python
#Calcul des corrélations entre variables
correlationsVariables = np.zeros((st.shape[1],st.shape[1]))
for k in range(st.shape[1]):
    correlationsVariables[:,k] = acp.components_[k,:] * sqrt_valeurspropres[k]
```

```python
np.set_printoptions(suppress=True)
print(correlationsVariables)
```

```python
print(pd.DataFrame({'id':df.columns[0:10],'COR_AXE_1':correlationsVariables[:,0],'COR_AXE_2':correlationsVariables[:,1]}))
```

```python
fig, axes = plt.subplots(figsize=(12,12))
axes.set_xlim(-1,1)
axes.set_ylim(-1,1)
for j in range(st.shape[1]): 
    plt.annotate(df.columns[j],(correlationsVariables[j,0],correlationsVariables[j,1]))
    plt.arrow(0,0,correlationsVariables[j,0]*0.9,correlationsVariables[j,1]*0.9, linestyle=':', capstyle="round", width=0.003, head_width=0.03)
plt.plot([-1,1],[0,0],color='silver',linestyle='-',linewidth=1)
plt.plot([0,0],[-1,1],color='silver',linestyle='-',linewidth=1)
cercle = plt.Circle((0,0),1,color='blue',fill=False)
axes.add_artist(cercle)
plt.show()
```
